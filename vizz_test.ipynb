```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/yuki-shi/pokedex-flask/main/pokemon.csv')
df.tail()
```

```python
df.corr()
```

```python
df2 = df
df2.loc[df2['Type 2'].isnull(), 'Type 2'] = df2['Type 1']
df2[df2['Type 2'].isnull()]
df2 = df.groupby(['Type 1', 'Type 2']).nunique()['No.'].reset_index()
df2 = df2.sort_values('No.', ascending=False)

fig = go.Figure(data=go.Heatmap(
                        z=df2['No.'],
                        x=df2['Type 1'],
                        y=df2['Type 2'],
                        hoverongaps=False,
                        colorscale='Viridis'))

fig.update_layout(
#    paper_bgcolor='rgba(0,0,0,0)',
#    plot_bgcolor='rgba(0,0,0,0)',
    height=550,
    width=550
)
```

```python
filter = 'grass'
color = 'rgba(53, 170, 114, 0.95)'
```

```python
df3 = df.loc[(df['Type 1'] == f'{filter}') | (df['Type 2'] == f'{filter}'), :]
df3.loc[df3['Type 2'].isnull(), 'Type 2'] = df3['Type 1']
df3 = df3.groupby(['Type 1', 'Type 2']).nunique()['No.'].reset_index()
df3.loc[df3['Type 1'] == df3['Type 2'], 'type-agg'] = df3['Type 1']
df3.loc[df3['Type 1'] != df3['Type 2'], 'type-agg'] = df3[['Type 1', 'Type 2']].transpose().agg(' + '.join)
df3.sort_values('No.', ascending=False, inplace=True)

fig = go.Figure(data=[go.Bar(x=df3['type-agg'], y=df3['No.'])])
fig.update_layout(width=750, height=550)
fig.update_traces(marker_color=color)
```

```python
df3 = df.loc[(df['Type 1'] == f'{filter}') | (df['Type 2'] == f'{filter}'), :]
df_cat = df3.select_dtypes(exclude='object')
df_cat = df_cat.drop('No.', axis=1)
df_stats = df_cat[['HP', 'Att', 'Def', 'S.Att', 'S.Def', 'Spd']]
```

```python
fig = go.Figure()

for column in df_stats.columns:
  fig.add_trace(go.Box(y=df_stats[column], name=column))

fig.update_layout(height=550,
                  width=550)
fig.show()
```
